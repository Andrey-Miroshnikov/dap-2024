Задание №7. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

````python
younger = titanic_df[titanic_df['Age'] < 30]
older = titanic_df[titanic_df['Age'] > 60]
ysurvived = round(younger['Survived'].mean(), 2)
osurvived = round(older['Survived'].mean(), 2)
print("Survived percentage for people younger than 30 y.o. is", ysurvived)
print("Survived percentage for people older than 60 y.o. is", osurvived)
print("Given statement is true.") if ysurvived > osurvived else print("Given statement is false.")
````

````python
male = titanic_df[titanic_df['Sex'] == "male"]
female = titanic_df[titanic_df['Sex'] == "female"]
msurvived = round(male['Survived'].mean(), 2)
fsurvived = round(female['Survived'].mean(), 2)
print("Survived percentage for men is", msurvived)
print("Survived percentage for women is", fsurvived)
print("Given statement is true.") if fsurvived > msurvived else print("Given statement is false.")
````

# Ответы на вопросы по pandas и статистике

## 1. Что такое медиана? Рассчитать медиану для предложенного ряда чисел.

**Медиана** - это значение, которое делит упорядоченный набор данных пополам. 50% значений находятся ниже медианы, и 50% - выше.

**Расчет медианы:**
Для ряда чисел: [12, 8, 15, 6, 10, 9, 11]
1. Упорядочиваем: [6, 8, 9, 10, 11, 12, 15]
2. Находим средний элемент: 10
**Медиана = 10**

## 2. Что такое дисперсия? Рассчитать дисперсию для предложенного ряда чисел.

**Дисперсия** - это среднее арифметическое квадратов отклонений значений от их среднего значения. Показывает меру разброса данных.

**Расчет дисперсии** для ряда: [12, 8, 15, 6, 10, 9, 11]
1. Среднее значение: (12+8+15+6+10+9+11)/7 = 71/7 = 10.14
2. Квадраты отклонений:
   (12-10.14)²=3.46, (8-10.14)²=4.58, (15-10.14)²=23.62,
   (6-10.14)²=17.14, (10-10.14)²=0.02, (9-10.14)²=1.30, (11-10.14)²=0.74
3. Сумма квадратов: 3.46+4.58+23.62+17.14+0.02+1.30+0.74 = 50.86
4. Дисперсия: 50.86/7 = 7.27
**Дисперсия = 7.27**

## 3. Что такое среднеквадратичное отклонение? Рассчитать для предложенного ряда чисел.

**Среднеквадратичное отклонение (стандартное отклонение)** - это квадратный корень из дисперсии. Показывает средний разброс значений относительно среднего.

**Расчет:**
Стандартное отклонение = √Дисперсия = √7.27 ≈ 2.70
**Среднеквадратичное отклонение ≈ 2.70**

## 4. Что такое Series и чем он отличается от DataFrame

**Series** - это одномерный помеченный массив, который может содержать данные любого типа. Похож на столбец в таблице.

**DataFrame** - это двумерная помеченная структура данных с столбцами потенциально разных типов. Похож на таблицу или электронную таблицу.

**Отличия:**
- Series: 1 измерение, DataFrame: 2 измерения
- Series: один тип данных для всех элементов, DataFrame: разные типы в разных столбцах
- Series: один индекс, DataFrame: индекс строк и имена столбцов

## 5. Можно ли изменить индекс в Series после создания объекта?

**Да**, можно изменить индекс после создания с помощью присвоения нового значения атрибуту index:
```python
series.index = new_index
```
## 6. Операции сложения над двумя объектами Series

- **Одинаковые типы данных, одинаковые размеры и одинаковые индексы?**  
  Да, сложение выполняется поэлементно

- **Одинаковые типы данных, одинаковые размеры, но разные индексы?**  
  Да, но результат будет содержать объединение индексов, отсутствующие значения = NaN

- **Одинаковые типы данных, разные размеры и одинаковые индексы?**  
  Да, для совпадающих индексов, остальные = NaN

- **Разные типы данных, одинаковые размеры и одинаковые индексы?**  
  Нет, если типы несовместимы (например, строка + число), либо будет приведение типов

- **Разные типы данных, одинаковые размеры и разные индексы?**  
  Нет, те же проблемы с типами + проблемы с индексами

- **Разные типы данных, разные размеры и разные индексы?**  
  Нет, максимальная несовместимость

## 7. Что такое DataFrame и чем он отличается от Series

**DataFrame** - это двумерная помеченная структура данных с столбцами потенциально разных типов, похожая на таблицу.

**Отличия от Series:**
- Размерность: DataFrame - 2D, Series - 1D
- Индексация: DataFrame имеет индекс строк и имена столбцов
- Типы данных: разные столбцы могут иметь разные типы
- Структура: DataFrame можно рассматривать как словарь Series

## 9. Чем отличается метод iloc от метода loc

**iloc** - индексация по целочисленным позициям (как в массивах NumPy)

```python
df.iloc[0]    # первая строка
df.iloc[0:5]  # строки с 0 по 4
```
**loc** - индексация по меткам индекса (по именам)

```python
df.loc['a']         # строка с индексом 'a'
df.loc['a':'c']     # строки от 'a' до 'c' включительно
```
