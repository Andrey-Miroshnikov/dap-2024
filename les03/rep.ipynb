Задача №3. Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8

````python
print("\n3. Смена знака у элементов от 3 до 8:")
z = np.arange(11)
print("Исходный массив:", z)
z[(z > 3) & (z < 8)] *= -1
print("Измененный массив:", z)
````

Задача №6. Дан массив 10x2 подеставлющий собой точки в декартовой системе координат, преобразовать в полярную

````python
print("\n6. Преобразование в полярные координаты:")
# Создаем массив с простыми целыми координатами
z = np.random.randint(-5, 6, (10, 2))
print("Декартовы координаты:")
print(z)
x, y = z[:, 0], z[:, 1]
r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)
polar_coords = np.column_stack((r, theta))
print("Полярные координаты:")
print(polar_coords)
````

Задача №10. Найти ближайшее к заданному значению число в заданном массиве

````python
print("\n10. Поиск ближайшего числа:")
z = np.arange(100)
v = np.random.uniform(0, 100)
print("Сгенерированный массив:")
print(z)
print(f"\nИскомое значение: {v:.2f}")
idx = (np.abs(z - v)).argmin()
nearest = z[idx]
print(f"Ближайшее число: {nearest}")
print(f"Индекс ближайшего числа: {idx}")
````

## Ответы на теоретические вопросы

### Вопрос 1: Какой тип (dtype) будут иметь элементы массива?

- `np.array([1, 4, 2, 5, 3])` → `int64` (целые числа по умолчанию)
- `np.array([1., 4, 2, 5, 3])` → `float64` (присутствует float, поэтому повышение до float)
- `np.array([1, 4, 2, 5, 3], dtype='float128')` → `float128` (явное указание типа)
- `np.array([1., 4, 2, 5, 3], dtype='complex128')` → `complex128` (явное указание комплексного типа)
- `np.array([1, 4, 2, 5, 3], dtype='float64')` → `float64` (явное указание типа)
- `np.array([1, 4, 2, 5, 3], dtype='int64')` → `int64` (явное указание типа)
- `np.array([1, 4, 2, 5, 3], dtype='bool')` → `bool` (0 → False, остальное → True)
- `np.array([1, 4, 2, 5, 3], dtype='str')` → `<U1` или `<U21` (строковый тип, длина зависит от платформы)
- `np.array([1, 4, 2, 5, '3'])` → `<U11` (смешанные типы → строка, самый общий тип)
- `np.array([1, -4, 2, 0, 1j])` → `complex128` (присутствует комплексное число)

### Вопрос 2: Что будет выведено в консоль и почему?

- `np.array([1, 4, 2, 5, 3], dtype='str')[1] + '1'` → `'41'`  
  **Почему**: Элемент с индексом 1 это строка '4', конкатенация с '1' дает '41'

- `np.array([1, -4, 2, 0, 3], dtype='bool')[3] or np.array([1, -4, 2, 0, 3])[1]` → `-4`  
  **Почему**: `dtype='bool'` преобразует числа: 0 → False, остальное → True. [3] = 0 → False. Операция `or` возвращает первый истинный операнд или последний. `np.array([1, -4, 2, 0, 3])[1] = -4` (True) → возвращается -4

- `np.array([1, -4, 2, 0, 3], dtype='bool')[3] and np.array([1, -4, 2, 0, 3])[1]` → `False`  
  **Почему**: [3] = 0 → False. Операция `and` возвращает первый ложный операнд → False

- `np.array([1, -4, 2, 0, True], dtype='int64')[-1]` → `1`  
  **Почему**: True преобразуется в целое число 1

- `np.array([1, -4, 2, 0, 1j], dtype='int64')[1]` → **Ошибка!**  
  **Почему**: Нельзя явно преобразовать комплексное число 1j в int64

- `np.array([1, -4, 2, 0, 1j], dtype='str')[-1]` → `'1j'`  
  **Почему**: Комплексное число преобразуется в строку '1j'

- `np.array([1, 4, 2, 5, '3'])[-1] * 3` → `'333'`  
  **Почему**: Элемент '3' это строка, умножение строки на число повторяет ее → '333'

### Вопрос 3: Каков результат операции?

- `np.ones((3, 5), dtype=float)` → Матрица 3x5 из единиц типа float64
- `np.full((3, 5), 3.14)` → Матрица 3x5, заполненная значением 3.14
- `np.linspace(0, 1, 5)` → `array([0. , 0.25, 0.5 , 0.75, 1. ])` (5 равномерно distributed точек)
- `np.diag((1,3,5))` → Диагональная матрица 3x3 с элементами 1, 3, 5 на диагонали
- `np.smile((1,3,5))` → **Ошибка!** Функция `np.smile()` не существует
- `np.random.random((3, 3))` → Матрица 3x3 со случайными числами от 0 до 1
- `np.eye(3)` → Единичная матрица 3x3
- `np.empty(3)` → Массив из 3 элементов без инициализации (значения могут быть любыми)

### Вопрос 4: Дан двумерный массив X, что будет выведено в консоль?

```python
X = np.array([[3, 5, 2, 4],
              [7, 6, 8, 8],
              [1, 6, 7, 7]])
```

- **X[0,0]** → 3 (элемент в 0 строке, 0 столбце)
- **X[-1,-1]** → 7 (элемент в последней строке, последнем столбце)
- **X[0:1,-1]** → [4] (срез: строка 0, последний столбец)
- **X[0:2,-1]** → [4 8] (строки 0 и 1, последний столбец)
- **X[0:2,0:2]** → [[3 5] [7 6]] (строки 0-1, столбцы 0-1)
- **X[1,1:3]** → [6 8] (строка 1, столбцы 1-2)
- **X[1,:]** → [7 6 8 8] (вся строка 1)
- **X[:,:]** → Вся матрица (аналог X)
- **X[:]** → Вся матрица (аналог X)
- **X[::2,::2]** → [[3 2] [1 7]] (каждая вторая строка и столбец)
- **X[0:3:2,0:3:2]** → [[3 2] [1 7]] (аналогично предыдущему)
- **X[::-1,::-1]** → [[7 7 6 1] [8 8 6 7] [4 2 5 3]] (обратный порядок строк и столбцов)

### Вопрос 5: Дан двумерный массив grid, в чем различее между операциями?
```python
grid = np.arange(16).reshape((4, 4))
```

- **np.split(grid, [2])** → Разделяет массив по строкам после индекса 2. Получится список из двух массивов: строк 0-1 и 2-3
- **np.split(grid, [2], axis=1)** → Разделяет массив по столбцам после индекса 2. Получится список из двух массивов: столбцов 0-1 и 2-3

### Вопрос 8: Дан массив X, каковы результаты операций?
 ```python
 M = np.array([[3, 5, 2, 4],
              [7, 6, 8, 8],
              [1, 6, 7, 7]])
```

- **np.sum(M)** → 58 (сумма всех элементов)
- **np.sum(M, axis=1)** → [14 29 21] (сумма по строкам)
- **np.sum(M, axis=0)** → [11 17 17 19] (сумма по столбцам)
- **np.sum(M, axis=2)** → Ошибка! Ось 2 не существует (массив 2D)
- **np.sum(M, 1)** → [14 29 21] (аналог axis=1)
- **sum(M)** → [11 17 17 19] (для 2D массивов эквивалентно sum(M, axis=0))
- **sum(M,1)** → Ошибка! Функция sum() не принимает второй аргумент в таком виде

### Вопрос 9: Дан массив X, что будет выведено в консоль?

Массив X: [1 2 3 4]

- **X > 3** → [False False False True] (поэлементное сравнение)
- **X < 3** → [True True False False] (поэлементное сравнение)
- **X[X < 3]** → [1 2] (булева индексация - элементы, где условие True)
- **X[X > 3]** → [4] (булева индексация)
- **X[X != 3]** → [1 2 4] (булева индексация)

### Вопрос 10: В чем преимущество массивов NumPy по сравнению с обычными списками? Чем обусловлено это преимущество?

**Преимущества NumPy**:
- Высокая производительность (в 10-100 раз быстрее операций с списками)
- Векторизованные операции - возможность выполнять математические операции над всем массивом без циклов
- Эффективное использование памяти - данные хранятся в непрерывных блоках памяти
- Многомерные массивы - удобная работа с матрицами и тензорами
- Широкий набор математических функций - оптимизированные реализации математических операций
- Интеграция с научными библиотеками - совместимость с pandas, scikit-learn, matplotlib и др.

**Причины высокой производительности**:
- Непрерывное хранение данных в памяти (в отличие от разрозненных объектов в списках)
- Векторизованные операции на C/Fortran - низкоуровневая оптимизация
- Отсутствие накладных расходов на проверку типов элементов при операциях
- Кэш-дружественная организация данных - эффективное использование кэша процессора
- Использование SIMD-инструкций процессора для параллельных вычислений
